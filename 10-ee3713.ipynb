### Bayes Rule

Describes the probability of an event, based on prior knowledge of conditions that might be related to the event. This theorem provides a principled way for calculating conditional probability. Not only is it a powerful tool within probability but it is also widely used in the field of machine learning.

### Bayes Formula



- P(A) = The probability of A occuring
- P(A|B) = The probability of A given B
- P(B) = The probability of B occuring
- P(B|A) = The probability of B given A

A simple example for this would be the probability of rainy day based on a cloudy morning. 
- 50% of rainy days start of cloudy P(B|A)
- 40% of days start cloudy P(B)
- 10% of days end up rainy P(A)


```python
# calculate P(A|B) given P(A), P(B|A), P(B)
def bayes_theorem(p_a, p_b, p_b_given_a):
    p_a_given_b =(p_b_given_a * p_a) / p_b
    return p_a_given_b
# P(A)
p_a = .1
# P(B)
p_b = 0.4
# P(B|A)
p_b_given_a = 0.5
# calculate P(A|B)
result = bayes_theorem(p_a, p_b, p_b_given_a)
# summarize
print('P(A|B) = %.3f%%' % (result * 100))
```

### Law of Total Probability

This formula is a fundamental rule relating marginal probabilities to conditional probabilities and expresses the total probability of an outcome via several distinct events. 




### Example


There are four bags that each contain 100 marbles
- Bag 1 has 50 red and 50 blue marbles
- Bag 2 has 70 red and 30 blue marbles
- Bag 3 has 20 red and 80 blue marbles
- Bag 4 has 40 red and 60 blue marbles

(A) will be the event that the chosen marble is red. B refers to the chosen back
- P(A|B1) = .50
- P(A|B2) = .70
- P(A|B3) = .20
- P(A|B4) = .45

Using the Law we can write P(A) = P(A|B1)P(B1) + P(A|B2)P(B2) + P(A|B3)P(B3) + P(A|B4)(B4)
- = (.50)1/4 + (.70)1/4 + (.20)1/4 + (.45)1/4 
- = .46
